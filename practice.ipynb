Create a temporary table temp_high_marks that stores students with marks > 80.

Query the temporary table to show only name and marks.

Add a new student to the temporary table.

Drop the temporary table manually.

```python
%%sql
CREATE TEMPORARY TABLE high_marks AS
SELECT name,section,marks
FROM students
WHERE marks>90;
```

```python
%%sql

SELECT * FROM  high_marks;
```

```python
%%sql
SHOW TABLES;
```

| Can you do operations before WITH? | No |
| Must it end with a query using it? | Yes |
| Can the last query be only SELECT? | No, it can also be INSERT, UPDATE, DELETE |

```python
%%sql
WITH high_scores AS(
SELECT * FROM high_marks
WHERE marks>=75)
SELECT * FROM high_scores;
```
