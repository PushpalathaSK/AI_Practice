Create a temporary table temp_high_marks that stores students with marks > 80.

Query the temporary table to show only name and marks.

Add a new student to the temporary table.

Drop the temporary table manually.

```python
%%sql
CREATE TEMPORARY TABLE high_marks AS
SELECT name,section,marks
FROM students
WHERE marks>90;
```

```python
%%sql

SELECT * FROM  high_marks;
```

| Can you do operations before WITH? | No |
| Must it end with a query using it? | Yes |
| Can the last query be only SELECT? | No, it can also be INSERT, UPDATE, DELETE |

```python
%%sql
WITH high_scores AS(
SELECT * FROM high_marks
WHERE marks>=75)
SELECT * FROM high_scores;
```

## Primary key:
Ensures each record is unique and not NULL.
Each table can have only one primary key.

```python
%%sql
CREATE TABLE student (
    student_id INT PRIMARY KEY,
    name VARCHAR(50)
);
```

## UNIQUE:
Ensures all values in a column are different (but allows one NULL).

```python
%%sql
ALTER TABLE first.employees ADD COLUMN(
    email VARCHAR(100) UNIQUE
);
```

```python
%%sql
ALTER TABLE uber.rides
ADD CONSTRAINT pk_driver UNIQUE (driver_id);
```

```python
%%sql
desc uber.rides; #not displayed as pk_driver for driver_id
SHOW INDEX FROM uber.rides; #To actually see the constraint name, use
```

## NOT NULL:
Ensures a column cannot have NULL values.

```python
%%sql
CREATE TABLE users (
    username VARCHAR(50) NOT NULL
);
```

## CHECK:
Ensures the values meet a specific condition.

```python
%%sql
CREATE TABLE first.products (
    price DECIMAL(10,2) CHECK (price > 0)
);
```

## DEFAULT:
Sets a default value when no value is given.

```python
%%sql
CREATE TABLE accounts (
    balance DECIMAL(10,2) DEFAULT 0.00
);
```

## AUTO_INCREMENT / SERIAL:
Automatically generates a unique value (mostly used for IDs).

```python
%%sql
CREATE TABLE books (
    book_id INT AUTO_INCREMENT PRIMARY KEY,
    title VARCHAR(100)
);
```

## *COMPOSITE KEY*
Combines multiple columns as a single primary key.

```python
%%sql
CREATE TABLE attendance (
    student_id INT,
    date DATE,
    PRIMARY KEY (student_id, date)
);
```
