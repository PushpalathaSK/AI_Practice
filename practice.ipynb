## Foreign Key
A foreign key is a column (or set of columns) in one table that creates a link to another table’s primary key.
It ensures the relationship between the two tables remains consistent.

(A foreign key in one table is a reference to the primary key in another.)

```python
%%sql

CREATE TABLE departments (
    dept_id INT PRIMARY KEY,
    dept_name VARCHAR(50)
);

CREATE TABLE employees (
    emp_id INT PRIMARY KEY,
    emp_name VARCHAR(50),
    dept_id INT
    
);
```


**INSERT**-You can only insert rows in the child table if the foreign key exists in the parent.

**UPDATE**-You can’t change or delete a parent key if child rows depend on it (unless you specify CASCADE rules).

**DELETE**-Deleting a parent row that has child references will cause an error unless cascading is enabled.

```python
%%sql
ALTER TABLE employees
ADD CONSTRAINT fk_dept
FOREIGN KEY (dept_id)
REFERENCES departments(dept_id)
ON DELETE CASCADE
ON UPDATE CASCADE;
```


**CASCADE**	- Automatically deletes/updates child rows when parent changes.

**SET NULL** - Sets the foreign key column to NULL when parent is deleted.

**RESTRICT** - Prevents deleting parent if child rows exist (default).

**NO ACTION** - Same as RESTRICT (default in MySQL).

## Aggregation func

```python
%%sql
SELECT SUM(amount)AS total_amount FROM customer WHERE transaction_status='Completed';
SELECT MAX(amount)AS max_amount FROM customer WHERE transaction_status='Completed';
SELECT MIN(amount)AS min_amount FROM customer WHERE transaction_status='Completed';
#count,avg
```

If your SELECT includes both aggregate and non-aggregate columns,
every non-aggregate column must appear in the GROUP BY clause

GROUP BY tells SQL how to split the table into mini-tables before applying aggregation.

```python


%%sql

SELECT login_device,SUM(amount) AS total_sale 
FROM customer
GROUP BY login_device;
```

HAVING is used to filter grouped (aggregate) results —
after the GROUP BY has been applied.


**WHERE** - Individual rows	- Filtering before grouping	-WHERE amount > 20000

**HAVING** - Groups (after GROUP BY)-Filtering after aggregation-HAVING SUM(amount) > 50000

```python
%%sql
SELECT name, SUM(amount) AS total_spent
FROM customer
GROUP BY name
HAVING SUM(amount) > 50000;
```

Conditional Logic in SQL — The CASE Statement

CASE ... WHEN ... THEN ... ELSE ... END.

```python
%%sql

SELECT name,amount,
    CASE
        WHEN amount>30000 THEN 'High value'
        WHEN amount BETWEEN 10000 AND 29999 THEN 'Medium value'
        ELSE 'Low value'
    END AS purchase_category
FROM customer;
```

```python
%%sql
SELECT 
    name,
    IF(amount > 20000, 'High', 'Low') AS purchase_type
FROM Customer;
```

```python
%%sql
SELECT name,product,transaction_status,
    CASE 
        WHEN transaction_status = 'Completed' AND amount > 20000 THEN 'Successful & High Value'
        WHEN transaction_status = 'Completed' THEN 'Successful'
        ELSE 'Not Successful'
    END AS transaction_type
FROM Customer;
```
